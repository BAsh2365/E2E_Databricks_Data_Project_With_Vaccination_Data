```python
from pyspark.sql import functions as F

src = "s3://vaccination-project-data-malaysia/raw/vax_malaysia/"

bronze = (spark.read
    .option("header", True)
    .option("inferSchema", False)
    .option("recursiveFileLookup", "true")
    .csv(src)
    .withColumn("_ingested_at", F.current_timestamp())
    .withColumn("_source_file", F.col("_metadata.file_path")))

bronze.write.format("delta").mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("workspace.vax_pipeline.bronze_vax_malaysia")

display(spark.table("workspace.vax_pipeline.bronze_vax_malaysia"))
```
